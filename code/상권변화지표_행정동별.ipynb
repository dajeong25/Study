```python
#자치구와 행정동 연결을 위한 파일 read
import pandas as pd
gu = pd.read_csv('서울생활이동데이터_서울자치구코드_20210907.csv', usecols=["시군구", "name"], encoding='cp949')
gu.info()
```

```python
gu.head()
```

```python
dong = pd.read_csv('서울생활이동데이터_서울행정동코드_20210907.csv', usecols=["시군구", '읍면동',"name"], encoding='cp949')
dong.info()
```

```python
dong.head()
```

```python
# gu & dong 합침
gu_dong = pd.merge(gu, dong, left_on='시군구', right_on='시군구')
gu_dong.head()
```

```python
gu_dong.info()
```

```python
#읍면동의 겹치는 신사동 이름변경 필요
for i in gu_dong[(gu_dong['name_y']=='신사동')&(gu_dong['읍면동']==1121068)].index:
  gu_dong.at[i, 'name_y'] = '신사동(관악구)'
```

```python
for i in gu_dong[(gu_dong['name_y']=='신사동')&(gu_dong['읍면동']==1123051)].index:
  gu_dong.at[i, 'name_y'] = '신사동(강남구)'
```

```python
 gu_dong[(gu_dong['name_y']=='신사동(강남구)')]  #변경완료
```

```python
#상권변화지표와 시군구 데이터 합침
biz_loc = pd.merge(biz_area_chg, gu_dong, left_on='행정동_코드_명', right_on='name_y')
biz_loc.head()
```

```python
#다중 인덱스(구&동 분류)로 데이터 재구조화
biz_pivot = biz_loc.pivot(index=['name_x','행정동_코드_명'], columns='기준_코드', values='상권_변화_지표_명')
biz_pivot.head()
```

```python
#자치구별로 빈도수가 가장 높은 상권_변화_지표_명 추출
biz_freq = biz_pivot.copy()
biz_freq = biz_freq.groupby('name_x').describe(include='object')
biz_freq.columns.names = ['기준코드', '통계']
```

```python
#불필요한 columns 제거
biz_freq = biz_freq.drop(['count', 'unique'], axis=1, level=1)
biz_freq
```

```python
import json
import pandas as pd
import requests
import folium
# 지도 시각화 참고 링크 : https://teddylee777.github.io/visualization/folium

# 서울 자치구역 json raw파일(githubcontent) 불러오기
r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content
seoul_geo = json.loads(c)
```

```python
#서울 자치구별 boundary 시각화
m = folium.Map(location=[37.559819, 126.963895], zoom_start=11)
folium.GeoJson(seoul_geo, name='지역구').add_to(m)
m
```

```python
grade = {'다이나믹':0, '정체':1, '상권축소':2, '상권확장':3}
grade = {'다이나믹':0, '정체':1, '상권축소':2, '상권확장':3}
m = folium.Map(location=[37.559819, 126.963895], zoom_start=11)
folium.Choropleth(geo_data=seoul_geo, 
                  data=biz_freq, 
                  columns=('name_x', '2019/Q3'), 
                  key_on = 'feature.properties.구',
                  fill_color=grade, 
                  legend_name='상권_변화_지표_명').add_to(m))
```
